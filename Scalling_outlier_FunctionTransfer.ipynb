```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('loan.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
q1=dataset['CoapplicantIncome'].quantile(0.25)
q3=dataset['CoapplicantIncome'].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
min_range=q1-(1.5*iqr)
max_range=q3+(1.5*iqr)
min_range, max_range
```

```python
dataset=dataset[dataset['CoapplicantIncome']<=max_range]
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset['CoapplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome1']=pd.DataFrame(ss.transform(dataset[['CoapplicantIncome']]))
```

```python
plt.subplot(1, 2, 1)
plt.title('Before')
sns.distplot(dataset['CoapplicantIncome'])

plt.subplot(1, 2, 2)
plt.title('After')
sns.distplot(dataset['CoapplicantIncome1'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome2']=pd.DataFrame(ms.transform(dataset[['CoapplicantIncome']]))
```

```python
plt.subplot(1, 2, 1)
plt.title('Before')
sns.distplot(dataset['CoapplicantIncome'])

plt.subplot(1, 2, 2)
plt.title('After')
sns.distplot(dataset['CoapplicantIncome2'])
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
ft.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome3']=ft.transform(dataset[['CoapplicantIncome']])
```

```python
plt.subplot(1, 2, 1)
plt.title('Before')
sns.distplot(dataset['CoapplicantIncome'])

plt.subplot(1, 2, 2)
plt.title('FncitonTransfer')
sns.distplot(dataset['CoapplicantIncome3'])
```
